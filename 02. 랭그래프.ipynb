### 2-3. ë„ˆë¹„ ìš°ì„  íƒìƒ‰
* BFS(Breadth-First Search): ëŒ€í‘œì ì¸ ê·¸ë˜í”„ íƒìƒ‰ ì•Œê³ ë¦¬ì¦˜
* ì •ì ë“¤ê³¼ ê°™ì€ ë ˆë²¨ì— ìˆëŠ” ë…¸ë“œë“¤(í˜•ì œ ë…¸ë“œë“¤)ì„ ë¨¼ì € ìš°ì„  íƒìƒ‰í•˜ëŠ” ë°©ì‹
* í•œ ë‹¨ê³„ì”© ë‚´ë ¤ê°€ë©´ì„œ í•´ë‹¹ ë…¸ë“œì™€ ê°™ì€ ë ˆë²¨ì— ìˆëŠ” ë…¸ë“œë“¤ì„ ë¨¼ì € ìˆœíšŒí•¨

````python
graph = dict()
graph['A'] = ['B', 'C']
graph['B'] = ['A', 'D']
graph['C'] = ['A', 'G', 'H', 'I']
graph['D'] = ['B', 'E', 'F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C']
graph['H'] = ['C']
graph['I'] = ['C', 'J']
graph['J'] = ['I']
````

### ë°©ë¬¸í–ˆë˜ ë¦¬ìŠ¤íŠ¸

A B C D G H I E F J

### ë°©ë¬¸í•´ì•¼í•  ë¦¬ìŠ¤íŠ¸
```
{'A': ['B', 'C'],
 'B': ['A', 'D'],
 'C': ['A', 'G', 'H', 'I'],
 'D': ['B', 'E', 'F'],
 'E': ['D'],
 'F': ['D'],
 'G': ['C'],
 'H': ['C'],
 'I': ['C', 'J'],
 'J': ['I']}
```

````python
def bfs(graph, start_node):
    visited, need_visit = list(), list()
    need_visit.append(start_node)

    while need_visit:
        node = need_visit.pop(0)
        if node not in visited:
            visited.append(node)
            need_visit.extend(graph[node])
    return visited
````

````python
bfs(graph, 'A')
````

````python
from langgraph.graph import StateGraph, START, END
from typing_extensions import TypedDict
````

````python
# ì…ë ¥ì„ ìœ„í•œ ìŠ¤í‚¤ë§ˆ ì •ì˜
class InputState(TypedDict):
    question: str


# ì¶œë ¥ì„ ìœ„í•œ ìŠ¤í‚¤ë§ˆ ì •ì˜
class OutputState(TypedDict):
    answer: str


# ì…ë ¥ê³¼ ì¶œë ¥ì„ í•©í•œ ì¢…í•© ìŠ¤í‚¤ë§ˆ ì •ì˜
class OverallState(InputState, OutputState):
    pass
````

````python
# ì…ë ¥ì„ ì²˜ë¦¬í•˜ê³  ë‹µë³€ì„ ìƒì„±í•˜ëŠ” ë…¸ë“œ ì •ì˜
def answer_node(state: InputState):
    return {"answer": "bye", "question": state["question"]} # ìƒíƒœ ì—…ë°ì´íŠ¸


graph_builder = StateGraph(OverallState, input_schema=InputState, output_schema=OutputState)
graph_builder.add_node(answer_node)  # ë‹µë³€ ë…¸ë“œ ì¶”ê°€
graph_builder.add_edge(START, "answer_node")  # ì‹œì‘ ì—£ì§€ ì¶”ê°€
graph_builder.add_edge("answer_node", END)  # ë ì—£ì§€ ì¶”ê°€
graph = graph_builder.compile()  # Compile the graph

# ì…ë ¥ invoke ë° ê²°ê³¼ ì¶œë ¥
print(graph.invoke({"question": "hi"}))
````

````python
from typing import Annotated
from typing_extensions import TypedDict

from langgraph.graph import StateGraph, START, END

from operator import add

class State(TypedDict):
    messages: Annotated[list[str], add]

graph = StateGraph(State)
````

````python
def chatbot(state: State):
    answer = "ì•ˆë…•í•˜ì„¸ìš”! ë¬´ì—‡ì„ ë„ì™€ë“œë¦´ê¹Œìš”?"
    print("Answer : ", answer)

    return {"messages": [answer]}

graph.add_node("chatbot", chatbot)
````

````python
graph.add_edge(START, "chatbot")
graph.add_edge("chatbot", END)
graph = graph.compile()
````
